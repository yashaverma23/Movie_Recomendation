# Corrwith()

```python
data.columns
```

```python
pivot=data.pivot_table(index='userId',columns='title',values='rating')
pivot
```

```python
pivot['(500) Days of Summer (2009)']
```

```python
#.corrwith use to find co-relation 
pivot.corrwith(pivot['(500) Days of Summer (2009)']).sort_values(ascending=False)
```

```python
pivot.corrwith(pivot['(500) Days of Summer (2009)']).sort_values(ascending=False).head(100)
```
